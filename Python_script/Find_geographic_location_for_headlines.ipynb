```python
import pandas as pd
from geonamescache import GeonamesCache 
```

```python
headline_data = pd.read_pickle('headline_ex1.pkl')
gc = GeonamesCache()
```

```python
# get dict item
def findDict (d, keyName):
    if not isinstance (d, dict): return None
    if keyName in d: return d [keyName]
    for subdict in d.values ():
        r = findDict (subdict, keyName)
        if r: return r
```

```python
def get_city_property(city,prop):
    city_info = gc.get_cities_by_name(city)
    for i in city_info:
        return findDict(i,prop)
```

```python
headline_data['latitude'] = headline_data['city'].apply(lambda row : get_city_property(row,'latitude'))
headline_data['longitude'] = headline_data['city'].apply(lambda row : get_city_property(row,'longitude'))
headline_data['countrycode'] = headline_data['city'].apply(lambda row : get_city_property(row,'countrycode'))
```
